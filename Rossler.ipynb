```python
#Importamos las bibliotecas
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
#Definimos las ecuaciones que describen el sistema dentro de una funci√≥n

def rossler(a, b, c, x0, y0, z0, t_max, dt):
    
    pasos = int(t_max / dt)
    x = np.zeros(pasos)
    y = np.zeros(pasos)
    z = np.zeros(pasos)
    
    x[0], y[0], z[0] = x0, y0, z0
    
    for i in range(1, pasos):
        dx = -y[i-1] - z[i-1]
        dy = x[i-1] + a * y[i-1]
        dz = b + z[i-1] * (x[i-1] - c)
        
        x[i] = x[i-1] + dx * dt
        y[i] = y[i-1] + dy * dt
        z[i] = z[i-1] + dz * dt
    
    return x, y, z


```

```python
#Graficamos en base a los parametros propuestos

a, b, c = 0.2, 0.2, 7.0
x0, y0, z0 = 1.0, 1.0, 1.0
t_max, dt = 200, 0.01

x, y, z = rossler(a, b, c, x0, y0, z0, t_max, dt)

fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.plot(x, y, z, lw=0.5)
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Atractor de R√∂ssler")
plt.show()
```

```python
# Obtener los rangos de valores
x_min, x_max = np.min(x), np.max(x)
y_min, y_max = np.min(y), np.max(y)
z_min, z_max = np.min(z), np.max(z)

print(f"Rango de x: [{x_min:.2f}, {x_max:.2f}]")
print(f"Rango de y: [{y_min:.2f}, {y_max:.2f}]")
print(f"Rango de z: [{z_min:.2f}, {z_max:.2f}]")
```
